```python
def categorize_income(income):
    try:
        if 0 <= income <= 30000:
            return 'E'
        elif 30001 <= income <= 50000:
            return 'D'
        elif 50001 <= income <= 200000:
            return 'C'
        elif 200001 <= income <= 1000000:
            return 'B'
        elif income >= 1000001:
            return 'A'
    except:
        pass
```

```python
data['total_income_category'] = data['total_income'].apply(categorize_income)
```

```python
data['purpose'].unique()
```

```python
def categorize_purpose(row):
    try:
        if 'автом' in row:
            return 'операции с автомобилем'
        elif 'жил' in row or 'недвиж' in row:
            return 'операции с недвижимостью'
        elif 'свад' in row:
            return 'проведение свадьбы'
        elif 'образов' in row:
            return 'получение образования'
    except:
        return 'нет категории'
```

```python
data['purpose_category'] = data['purpose'].apply(categorize_purpose)
```

```python
def children_category(children):   #создадим функцию распределения количества детей у клиентов
    if children == 0:
        return 'нет детей'
    if children == 1:
        return 'один ребенок'
    if children == 2:
        return 'два ребенка'
    if children == 3:
        return 'три ребенка'
    if children >= 3:
        return 'многодетная семья' 
```

```python
data['childrens'] = data['children'].apply(children_category) #создали новый столбец childrens с информацией о количестве детей у клиентов
```

```python
data.head(10) #вывелем первые 10 строк, чтобы убедиться, что добавили новый столбец
```

```python
childrens = data.pivot_table(index = 'childrens', columns='debt', values='gender', aggfunc = 'count') #построим сводную таблицу
```

```python
childrens #выведем таблицу, чтобы проверить, что получилось
```

```python
childrens.columns = ['no_debt', 'debt'] #создаем столбцы есть зодолжность или нет
```

```python
childrens #выведем таблицу, чтобы проверить, что получилось
```

```python
childrens['procent_doljnikov'] = (childrens['debt'] / (childrens['debt'] + childrens['no_debt'])).map('{:.1%}'.format) #создадим новый столбец и посчитаем в ней долю должникол в процентах
```

```python
childrens.sort_values(by='procent_doljnikov', ascending=False) #отсортируем таблицу по убыванию
```
